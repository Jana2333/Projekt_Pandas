```python
# vzdalenost mezi stanicemi
from geopy.distance import geodesic
from itertools import combinations
@np.vectorize
def vzdalenost(lat1, lon1, lat2, lon2):
    rs = geodesic( (lat1, lon1), (lat2, lon2) ).kilometers
    return rs


dv = vzdalenost(spoj_pro_vypocet['start_station_latitude'].to_numpy(), spoj_pro_vypocet['start_station_longitude'].to_numpy(),
                  spoj_pro_vypocet['end_station_latitude'].to_numpy(), spoj_pro_vypocet['end_station_longitude'].to_numpy(), )

spoj_pro_vypocet["vzdalenost_v_km"] = dv.round(2)
spoj_pro_vypocet=spoj_pro_vypocet[["start_station_name","end_station_name","vzdalenost_v_km"]]
spoj_pro_vypocet.round(2).head()

```
